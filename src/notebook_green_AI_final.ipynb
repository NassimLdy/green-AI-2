- **Remplacement automatique du fond des images en fond de sable**

Afin de permettre au robot de s'entra√Æner √† d√©tecter au mieux des d√©chets sur la plage, nous avons remplac√© le fond de toutes les images en fond de sable.

```python
# Images de sable disponibles
sand_images = [
    f for f in os.listdir(ASSET_DIR)
    if f.lower().endswith((".jpg", ".jpeg", ".png"))
]

def replace_background_with_sand(img_path, save_path):
    """D√©toure l'objet avec GrabCut et le place sur un fond de sable."""

    img = cv2.imread(img_path)
    if img is None:
        return

    h, w = img.shape[:2]
    mask = np.zeros((h, w), np.uint8)

    rect = (10, 10, w - 20, h - 20)  # zone suppos√©e contenir l'objet
    bgModel = np.zeros((1, 65), np.float64)
    fgModel = np.zeros((1, 65), np.float64)

    try:
        cv2.grabCut(img, mask, rect, bgModel, fgModel, 5, cv2.GC_INIT_WITH_RECT)
    except:
        cv2.imwrite(save_path, img)
        return

    mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype("uint8")

    sand = cv2.imread(os.path.join(ASSET_DIR, random.choice(sand_images)))
    sand = cv2.resize(sand, (w, h))

    result = sand.copy()
    result[mask2 == 1] = img[mask2 == 1]  # colle uniquement l'objet

    cv2.imwrite(save_path, result)
```

Cette fonction d√©toure approximativement chaque d√©chet avec GrabCut et le recolle sur un fond de sable r√©aliste afin d‚Äôobtenir un dataset coh√©rent avec la simulation Pygame.

- **Application de la fonction**

```python
# G√©n√©ration du dataset : remplacement du fond par du sable
for cls in tqdm(classes, desc="Fond sable"):

    src_dir = os.path.join(DATASET_ORIG, cls)         # images originales
    dst_dir = os.path.join(DATASET_SAND, cls)         # images transform√©es
    os.makedirs(dst_dir, exist_ok=True)

    for img_name in os.listdir(src_dir):
        if not img_name.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        src_path = os.path.join(src_dir, img_name)
        dst_path = os.path.join(dst_dir, img_name)

        # applique le d√©tourage + collage sur sable
        if not os.path.exists(dst_path):
            replace_background_with_sand(src_path, dst_path)

print("Dataset RealWaste avec fond de sable g√©n√©r√©")
```

Cet affichage permet de v√©rifier visuellement la qualit√© du pr√©traitement : pr√©sence correcte du fond de sable, d√©coupe coh√©rente de l‚Äôobjet, diversit√© visuelle suffisante avant l‚Äôentra√Ænement PyTorch.

## 3. **Cr√©ation du jeu de donn√©es train et test** 

Nous s√©parons ici le dataset enrichi (dataset_sand) en deux ensembles distincts, un jeu d‚Äôentra√Ænement et un jeu de test, afin de pouvoir entra√Æner le mod√®le tout en √©valuant sa capacit√© √† g√©n√©raliser sur des images jamais vues. Les dossiers train et test seront dans le dossier dataset.

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# -----------------------------------------------
# 1) Dossiers train / test bas√©s sur dataset_sand
# -----------------------------------------------

TRAIN_DIR = os.path.join(DATA_DIR, "train")
TEST_DIR  = os.path.join(DATA_DIR, "test")

# On repart de z√©ro pour √™tre s√ªr que le split correspond bien √† dataset_sand
shutil.rmtree(TRAIN_DIR, ignore_errors=True)
shutil.rmtree(TEST_DIR, ignore_errors=True)
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(TEST_DIR, exist_ok=True)

print("üìÅ TRAIN_DIR :", TRAIN_DIR)
print("üìÅ TEST_DIR  :", TEST_DIR)

# -------------------------------------
# 2) Classes prises depuis dataset_sand
# -------------------------------------

classes = sorted([
    c for c in os.listdir(DATASET_SAND)
    if os.path.isdir(os.path.join(DATASET_SAND, c))
])

print("Classes trouv√©es dans dataset_sand :", classes)

# -------------------------------------------
# 3) Split train / test et copie des fichiers
# -------------------------------------------

test_ratio = 0.2  # 20% test, 80% train

for cls in classes:
    src_dir = os.path.join(DATASET_SAND, cls)
    imgs = [
        f for f in os.listdir(src_dir)
        if f.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    if len(imgs) == 0:
        print(f"Aucun fichier image pour la classe {cls}, ignor√©e.")
        continue

    train_imgs, test_imgs = train_test_split(
        imgs, test_size=test_ratio, random_state=42
    )

    train_cls_dir = os.path.join(TRAIN_DIR, cls)
    test_cls_dir  = os.path.join(TEST_DIR, cls)
    os.makedirs(train_cls_dir, exist_ok=True)
    os.makedirs(test_cls_dir, exist_ok=True)

    for f in train_imgs:
        shutil.copy2(os.path.join(src_dir, f),
                     os.path.join(train_cls_dir, f))

    for f in test_imgs:
        shutil.copy2(os.path.join(src_dir, f),
                     os.path.join(test_cls_dir, f))

print("Split train/test √† partir de dataset_sand termin√©.")
```

### 4. **Pr√©paration du dataset pour l‚Äôentra√Ænement : transformations, split train/validation et DataLoaders**

Ce prochain bloc configure d‚Äôabord la taille des images et la taille de batch, puis d√©finit deux pipelines de transformation :  
un pour l‚Äôentra√Ænement, incluant des augmentations simples (flip horizontal, rotation) afin de rendre le mod√®le plus robuste aux variations naturelles des d√©chets sur la plage,  
et un pour le test, volontairement plus neutre pour garantir une √©valuation fid√®le.  
Les images sont redimensionn√©es, converties en tenseurs et normalis√©es selon les statistiques d‚ÄôImageNet, ce qui est essentiel pour les mod√®les pr√©-entra√Æn√©s comme MobileNetV2 et ResNet18 
Le dataset d‚Äôentra√Ænement est ensuite s√©par√© en un ensemble *train* et *validation* afin de suivre la progression r√©elle du mod√®le pendant l‚Äôapprentissage.  
Les DataLoaders sont cr√©√©s pour charger efficacement les images par batch, avec m√©lange al√©atoire uniquement sur l‚Äôentra√Ænement.  
Enfin, le script affiche la taille de chaque split ainsi que les classes d√©tect√©es, et d√©duit automatiquement le nombre total de classes √† pr√©dire.

- **D√©finition des transformations d‚Äôentra√Ænement et de test (pr√©traitement)**

```python
IMG_SIZE = 224
BATCH_SIZE = 32 

train_transform = transforms.Compose([ 
    transforms.Resize((IMG_SIZE, IMG_SIZE)), 
    transforms.RandomHorizontalFlip(), 
    transforms.RandomRotation(20), 
    transforms.ToTensor(), 
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]) 
]) 
test_transform = transforms.Compose([ 
    transforms.Resize((IMG_SIZE, IMG_SIZE)), 
    transforms.ToTensor(), 
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]) 
]) 
```

- **Pr√©paration des ensembles d‚Äôentra√Ænement, validation et test**

```python
# Chargement du dataset d‚Äôapprentissage
full_train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform) 

# Cr√©ation du split train/validation
val_size = int(0.2 * len(full_train_dataset)) 
train_size = len(full_train_dataset) - val_size 

train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size]) 

# Chargement du dataset de test
test_dataset = datasets.ImageFolder(TEST_DIR, transform=test_transform) 

# Construction des DataLoaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True) 
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False) 
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False) 

# V√©rification des tailles et des classes
print("Train:", len(train_dataset)," Val:", len(val_dataset)," Test:", len(test_dataset)) 
print("Classes :", full_train_dataset.classes) 
num_classes = len(full_train_dataset.classes)
```

## 6. **Construction et comparaison de diff√©rents mod√®les de classification**

Dans cette section, nous mettons en place plusieurs architectures de r√©seaux de neurones destin√©es √† la classification des d√©chets : un CNN simple, MobileNetV2 et ResNet18. L‚Äôobjectif est d‚Äôentra√Æner ces mod√®les dans des conditions identiques afin d‚Äôobserver leurs diff√©rences en termes de comportement, de complexit√© et de co√ªt computationnel. Nous int√©grerons √©galement une estimation de leur empreinte carbone lors de l‚Äôentra√Ænement, conform√©ment aux principes de la Green AI. Cette approche nous permettra d‚Äôanalyser comment la structure d‚Äôun mod√®le influence √† la fois les performances de classification et son impact environnemental.

### 6.1. **CNN Simple**

- **D√©finition du mod√®le CNN simple**

Ce premier mod√®le est un CNN simple compos√© de trois blocs convolution‚Äìactivation‚Äìpooling, suivi d‚Äôun classifieur enti√®rement connect√©. Il sert de point de d√©part pour comparer ensuite des architectures plus avanc√©es.

```python
class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64,128,3,padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*28*28,256), nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self,x):
        x = self.features(x)
        return self.classifier(x)

cnn = SimpleCNN(num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(cnn.parameters(), lr=1e-3)

cnn
```

- **Boucle d‚Äôentra√Ænement et d‚Äô√©valuation du CNN**

Nous d√©finissons ici les fonctions permettant d‚Äôentra√Æner le mod√®le sur un epoch complet ainsi que la proc√©dure d‚Äô√©valuation sur l‚Äôensemble de validation. Cela permet de suivre l‚Äô√©volution de la performance et de la perte au fil des epochs.

```python
def train_epoch(model, loader):
    model.train()
    total, correct, loss_sum = 0,0,0

    for X,y in loader:
        X,y = X.to(device), y.to(device)
        optimizer.zero_grad()

        outputs = model(X)
        loss = criterion(outputs,y)
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs,1)
        correct += (preds==y).sum().item()
        total += X.size(0)
        loss_sum += loss.item()*X.size(0)

    return loss_sum/total, correct/total

def eval_epoch(model, loader):
    model.eval()
    total, correct, loss_sum = 0,0,0

    with torch.no_grad():
        for X,y in loader:
            X,y = X.to(device), y.to(device)
            outputs = model(X)
            loss = criterion(outputs,y)

            _, preds = torch.max(outputs,1)
            correct += (preds==y).sum().item()
            total += X.size(0)
            loss_sum += loss.item()*X.size(0)

    return loss_sum/total, correct/total


history = {"train_acc":[], "val_acc":[], "train_loss":[], "val_loss":[]}

for epoch in range(1,16):
    tr_loss, tr_acc = train_epoch(cnn, train_loader)
    val_loss, val_acc = eval_epoch(cnn, val_loader)

    history["train_loss"].append(tr_loss)
    history["train_acc"].append(tr_acc)
    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)

    print(f"[{epoch}/15] Train Acc={tr_acc:.3f} | Val Acc={val_acc:.3f}")
```

- **Visualisation de l‚Äô√©volution de la perte et de l‚Äôaccuracy**

Nous tra√ßons ici les courbes de perte et d‚Äôaccuracy pour l‚Äôentra√Ænement et la validation afin d‚Äôobserver le comportement du mod√®le au fil des epochs.

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history["train_loss"], label="Train")
plt.plot(history["val_loss"], label="Val")
plt.title("Loss"); plt.legend()

plt.subplot(1,2,2)
plt.plot(history["train_acc"], label="Train")
plt.plot(history["val_acc"], label="Val")
plt.title("Accuracy"); plt.legend()

plt.show()
```

**Analyse des performances du CNN simple**

Les courbes montrent une diminution nette de la loss d‚Äôentra√Ænement, signe que le mod√®le parvient √† apprendre progressivement les motifs li√©s aux diff√©rents d√©chets.  
La loss de validation baisse √©galement au d√©but mais devient plus irr√©guli√®re ensuite, indiquant que le mod√®le l√©ger atteint ses limites et peine √† g√©n√©raliser parfaitement.  
Du c√¥t√© de l‚Äôaccuracy, l‚Äôentra√Ænement progresse de mani√®re r√©guli√®re tandis que la validation suit une trajectoire similaire mais avec un plafond plus bas.  
L‚Äô√©cart qui se creuse entre les deux courbes traduit un d√©but de surapprentissage : le mod√®le m√©morise mieux ses donn√©es d‚Äôentra√Ænement qu‚Äôil ne s‚Äôadapte √† de nouvelles images.  
Malgr√© cela, l‚Äôaccuracy en validation reste correcte pour un r√©seau aussi simple, surtout face √† un dataset visuellement vari√© et difficile comme celui des d√©chets sur plage.  
En r√©sum√©, ce CNN apprend efficacement mais montre rapidement ses limites structurelles, ce qui justifie le passage √† un mod√®le plus avanc√© comme MobileNetV2 pour aller plus loin.


- **√âvaluation du mod√®le sur le jeu de test : m√©triques et matrice de confusion**

Nous √©valuons ici les performances finales du mod√®le sur le jeu de test ind√©pendant. Le rapport de classification permet d‚Äôobtenir la pr√©cision, le recall et le F1-score pour chaque classe, tandis que la matrice de confusion met en √©vidence les erreurs typiques du mod√®le.

### 6.2. **Mobile netV2**

Apr√®s avoir test√© un CNN simple, nous passons maintenant √† MobileNetV2, un mod√®le bien plus performant et optimis√© pour la vision sur appareils l√©gers.  
Sa structure en *inverted residuals* et *depthwise separable convolutions* lui permet d‚Äôextraire des caract√©ristiques beaucoup plus riches tout en restant rapide et peu co√ªteux en calcul.  
Ce choix est particuli√®rement adapt√© √† notre contexte : d√©tecter la nature des d√©chets sur une plage, potentiellement en conditions embarqu√©es (robot, drone, mobile‚Ä¶).  
L‚Äôentra√Ænement qui suit vise donc √† exploiter cette architecture avanc√©e pour am√©liorer nettement la pr√©cision tout en conservant une bonne capacit√© de g√©n√©ralisation.  
Nous allons pr√©parer MobileNetV2, ajuster sa derni√®re couche au nombre de classes, puis lancer une phase d‚Äôapprentissage compl√®te sur notre dataset sable + d√©chets.

- **Construction du mod√®le MobileNetV2 adapt√© au nombre de classes**

Dans cette cellule, nous cr√©ons notre version de MobileNetV2. On part du mod√®le pr√©-entra√Æn√© sur ImageNet, puis on remplace simplement sa derni√®re couche pour qu‚Äôil pr√©dise nos cinq classes de d√©chets au lieu des 1000 classes d‚Äôorigine.

```python
import torch
import torch.nn as nn
from torchvision import models

def build_mobilenet_v2(num_classes):
    # Charger MobileNetV2 pr√©-entra√Æn√©
    model = models.mobilenet_v2(weights=models.MobileNet_V2_Weights.IMAGENET1K_V1)

    # On r√©cup√®re la taille de l‚Äôentr√©e du dernier classifier
    in_features = model.classifier[1].in_features

    # Remplacer la derni√®re couche par notre propre classifier
    model.classifier[1] = nn.Linear(in_features, num_classes)

    return model
```

- **Initialisation de MobileNetV2, de la fonction de perte et de l‚Äôoptimiseur**

On instancie ici MobileNetV2 avec notre nouvelle couche de sortie, puis on d√©finit la fonction de perte et l‚Äôoptimiseur qui serviront pendant l‚Äôentra√Ænement. Comme MobileNetV2 est d√©j√† pr√©-entra√Æn√©, on utilise un taux d‚Äôapprentissage plus faible pour √©viter de d√©stabiliser les poids appris sur ImageNet.

```python
mobilenet = build_mobilenet_v2(num_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(mobilenet.parameters(), lr=1e-4)  # lr plus petit !
```

- **D√©finition des transformations sp√©cifiques √† MobileNetV2**

MobileNetV2 attend des images pr√©trait√©es selon les statistiques d‚ÄôImageNet. Nous d√©finissons donc ici les transformations utilis√©es pendant l‚Äôentra√Ænement (avec augmentations) et celles utilis√©es pour l‚Äô√©valuation (sans augmentation), en conservant le m√™me format d‚Äôentr√©e que le mod√®le pr√©-entra√Æn√©.

```python
mobilenet_transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])

mobilenet_transform_eval = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])
```

- **Boucle d‚Äôentra√Ænement et d‚Äô√©valuation pour MobileNetV2**

Cette cellule contient la boucle compl√®te d‚Äôentra√Ænement et de validation pour MobileNetV2. Le mod√®le est entra√Æn√© pendant plusieurs epochs, et nous enregistrons √† chaque √©tape la perte et l‚Äôaccuracy, afin de suivre sa progression et de comparer ses performances au CNN simple.

```python
num_epochs = 10
history_mob = {
    "train_loss": [], "train_acc": [],
    "val_loss": [], "val_acc": []
}

for epoch in range(num_epochs):
    # TRAIN
    mobilenet.train()
    running_loss, running_corrects, total = 0, 0, 0

    for X, y in train_loader:
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = mobilenet(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs, 1)
        running_loss += loss.item() * X.size(0)
        running_corrects += (preds == y).sum().item()
        total += X.size(0)

    train_loss = running_loss / total
    train_acc = running_corrects / total

    # VAL
    mobilenet.eval()
    val_loss, val_corrects, val_total = 0, 0, 0

    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(device), y.to(device)
            outputs = mobilenet(X)
            loss = criterion(outputs, y)

            _, preds = torch.max(outputs, 1)
            val_loss += loss.item() * X.size(0)
            val_corrects += (preds == y).sum().item()
            val_total += X.size(0)

    val_loss = val_loss / val_total
    val_acc = val_corrects / val_total

    history_mob["train_loss"].append(train_loss)
    history_mob["train_acc"].append(train_acc)
    history_mob["val_loss"].append(val_loss)
    history_mob["val_acc"].append(val_acc)

    print(f"[{epoch+1}/{num_epochs}] Train Acc={train_acc:.3f} | Val Acc={val_acc:.3f}")
```

Nous tra√ßons ici l‚Äô√©volution de la perte et de l‚Äôaccuracy de MobileNetV2 sur les ensembles d‚Äôentra√Ænement et de validation, afin d‚Äôobserver comment le mod√®le apprend au fil des epochs et d‚Äôidentifier d‚Äô√©ventuels signes d‚Äôoverfitting.

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history_mob["train_loss"], label="Train")
plt.plot(history_mob["val_loss"], label="Val")
plt.title("Loss"); plt.legend()

plt.subplot(1,2,2)
plt.plot(history_mob["train_acc"], label="Train")
plt.plot(history_mob["val_acc"], label="Val")
plt.title("Accuracy"); plt.legend()

plt.show()
```

### 6.3. **ResNet18**

Apr√®s avoir travaill√© avec MobileNetV2, un mod√®le l√©ger optimis√© pour l‚Äôefficacit√© √©nerg√©tique, nous explorons maintenant une architecture plus classique et plus lourde : **ResNet18**.  
ResNet est devenu un standard en vision, notamment gr√¢ce aux *skip connections* qui facilitent l‚Äôapprentissage de r√©seaux profonds.

L‚Äôobjectif de cette section est double :  
- mesurer ce que l‚Äôon gagne r√©ellement en pr√©cision avec un mod√®le plus massif,  
- analyser le **surco√ªt √©nerg√©tique** associ√©, dans une d√©marche Green AI.

Nous allons donc charger ResNet18 pr√©-entra√Æn√©, l‚Äôadapter √† notre t√¢che, et comparer ses performances et son empreinte carbone √† celles de MobileNetV2.

- **Construction du mod√®le ResNet18**

On introduit ici un second mod√®le : plus profond que MobileNetV2 et souvent plus performant.
L‚Äôobjectif est d‚Äôadapter ResNet18 √† notre propre t√¢che de classification en rempla√ßant sa derni√®re couche.

```python
import torch
import torch.nn as nn
from torchvision import models

def build_resnet18(num_classes):
    # Charger ResNet18 pr√©-entra√Æn√© sur ImageNet
    model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
    
    # R√©cup√©rer la taille de l'entr√©e de la derni√®re couche fully-connected
    in_features = model.fc.in_features
    
    # Remplacer la derni√®re couche par notre propre classifier
    model.fc = nn.Linear(in_features, num_classes)
    
    return model
```

- **Initialisation du mod√®le ResNet18 et param√©trage de l‚Äôoptimisation**

On instancie maintenant ResNet18 adapt√© √† notre nombre de classes, puis on configure la fonction de perte et l‚Äôoptimiseur.
Le learning rate reste volontairement faible afin de fine-tuner le r√©seau sans perturber ses poids pr√©-entra√Æn√©s.

```python
resnet = build_resnet18(num_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(resnet.parameters(), lr=1e-4)  # petit lr pour le fine-tuning
```

- **Pr√©paration des transformations d‚Äôimages pour ResNet18**

Pour exploiter pleinement les capacit√©s de ResNet18, nous appliquons un pr√©traitement coh√©rent avec son entra√Ænement initial (ImageNet).
Les images sont redimensionn√©es, normalis√©es et enrichies par des augmentations pour renforcer la robustesse du mod√®le.

```python
from torchvision import transforms

resnet_transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])

resnet_transform_eval = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])
```

- **Entra√Ænement et validation du mod√®le ResNet18**

Nous lan√ßons maintenant la phase d‚Äôapprentissage du r√©seau.
L‚Äôobjectif : ajuster progressivement les poids du mod√®le via le fine-tuning, mesurer sa progression √† chaque √©poque et surveiller sa capacit√© de g√©n√©ralisation gr√¢ce au suivi des performances sur le jeu de validation.

```python
num_epochs = 10
history_res = {
    "train_loss": [], "train_acc": [],
    "val_loss": [], "val_acc": []
}

for epoch in range(num_epochs):
    # ------- TRAIN -------
    resnet.train()
    running_loss, running_corrects, total = 0, 0, 0

    for X, y in train_loader:
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = resnet(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs, 1)
        running_loss     += loss.item() * X.size(0)
        running_corrects += (preds == y).sum().item()
        total            += X.size(0)

    train_loss = running_loss / total
    train_acc  = running_corrects / total

    # ------- VAL -------
    resnet.eval()
    val_loss, val_corrects, val_total = 0, 0, 0

    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(device), y.to(device)
            outputs = resnet(X)
            loss = criterion(outputs, y)

            _, preds = torch.max(outputs, 1)
            val_loss      += loss.item() * X.size(0)
            val_corrects  += (preds == y).sum().item()
            val_total     += X.size(0)

    val_loss = val_loss / val_total
    val_acc  = val_corrects / val_total

    history_res["train_loss"].append(train_loss)
    history_res["train_acc"].append(train_acc)
    history_res["val_loss"].append(val_loss)
    history_res["val_acc"].append(val_acc)

    print(f"[{epoch+1}/{num_epochs}] Train Acc={train_acc:.3f} | Val Acc={val_acc:.3f}")
```

- **Visualisation de l‚Äôapprentissage : √©volution de la loss et de l‚Äôaccuracy**

Pour mieux comprendre le comportement du mod√®le au fil des √©poques, nous retra√ßons ici les courbes de perte et de pr√©cision.
Ces graphiques permettent d‚Äôidentifier rapidement un surapprentissage √©ventuel, une convergence stable ou au contraire des difficult√©s d‚Äôoptimisation.

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history_res["train_loss"], label="Train")
plt.plot(history_res["val_loss"], label="Val")
plt.title("Loss"); plt.legend()

plt.subplot(1,2,2)
plt.plot(history_res["train_acc"], label="Train")
plt.plot(history_res["val_acc"], label="Val")
plt.title("Accuracy"); plt.legend()

plt.show()
```

**Analyse des courbes d‚Äôentra√Ænement de ResNet18 :**

La loss d‚Äôentra√Ænement chute rapidement, signe que ResNet18 exploite efficacement ses blocs r√©siduels pour apprendre des repr√©sentations visuelles pertinentes.  
La loss de validation descend au d√©but puis remonte l√©g√®rement : cela sugg√®re un d√©but de surapprentissage, sans √™tre encore critique.  
C√¥t√© accuracy, ResNet18 atteint tr√®s vite un haut niveau de performance en entra√Ænement, d√©passant les 95 %.  
L‚Äôaccuracy de validation progresse bien mais plafonne plus t√¥t, ce qui refl√®te une capacit√© de g√©n√©ralisation un peu moins bonne que celle du MobileNetV2 observ√©e pr√©c√©demment.  
L‚Äô√©cart entre train et val augmente progressivement, signe que le mod√®le devient trop sp√©cialis√© sur son jeu d‚Äôentra√Ænement.

- **Suivi des √©missions carbone lors de l‚Äôentra√Ænement de ResNet18**
  
Pour √©valuer l‚Äôimpact √©nerg√©tique de notre approche, nous instrumentons l‚Äôentra√Ænement de ResNet18 avec CodeCarbon. Ce suivi permet de quantifier pr√©cis√©ment la consommation √©lectrique et les √©missions associ√©es √† chaque epoch, afin de comparer objectivement le co√ªt environnemental des architectures test√©es.


```python
from codecarbon import EmissionsTracker
models_dir = os.path.join("..", "models")
num_epochs = 10
history_res = {
    "train_loss": [], "train_acc": [],
    "val_loss": [], "val_acc": []
}

emissions_file_res = "emissions_resnet18_sand.csv"

tracker_res = EmissionsTracker(
    output_dir=models_dir,                       # ../models
    output_file=emissions_file_res,
    project_name="ResNet18_sand_training",
    on_csv_write="update"
)

tracker_res.start()   #  d√©but mesure CO‚ÇÇ

for epoch in range(num_epochs):
    # --------- TRAIN ---------
    resnet.train()
    running_loss, running_corrects, total = 0, 0, 0

    for X, y in train_loader:
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = resnet(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs, 1)
        running_loss     += loss.item() * X.size(0)
        running_corrects += (preds == y).sum().item()
        total            += X.size(0)

    train_loss = running_loss / total
    train_acc  = running_corrects / total

    # --------- VALIDATION ---------
    resnet.eval()
    val_loss, val_corrects, val_total = 0, 0, 0

    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(device), y.to(device)
            outputs = resnet(X)
            loss = criterion(outputs, y)

            _, preds = torch.max(outputs, 1)
            val_loss      += loss.item() * X.size(0)
            val_corrects  += (preds == y).sum().item()
            val_total     += X.size(0)

    val_loss = val_loss / val_total
    val_acc  = val_corrects / val_total

    history_res["train_loss"].append(train_loss)
    history_res["train_acc"].append(train_acc)
    history_res["val_loss"].append(val_loss)
    history_res["val_acc"].append(val_acc)

    print(
        f"[ResNet18 {epoch+1}/{num_epochs}] "
        f"Train Loss={train_loss:.4f} Acc={train_acc:.3f} | "
        f"Val Loss={val_loss:.4f} Acc={val_acc:.3f}"
    )

    tracker_res.flush()   # ‚¨Ö enregistre les mesures CO‚ÇÇ de cette epoch

tracker_res.stop()
print(" Suivi CO‚ÇÇ termin√© pour ResNet18.")
print(f" Fichier CO‚ÇÇ g√©n√©r√© dans models/ : {emissions_file_res}")
```

- **Chargement des mesures d‚Äôempreinte carbone**

Pour comparer l‚Äôimpact environnemental de MobileNetV2 et ResNet18, nous commen√ßons par charger les fichiers CSV g√©n√©r√©s par CodeCarbon.  
Ce bloc lit automatiquement les donn√©es, d√©tecte la colonne contenant les √©missions et pr√©pare les tableaux pour les √©tapes de comparaison et de visualisation.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt

models_dir = os.path.join("..", "models")

file_mobilenet = os.path.join(models_dir, "emissions_mobilenetv2_sand.csv")
file_resnet    = os.path.join(models_dir, "emissions_resnet18_sand.csv")

def load_emissions_csv(filepath):
    if not os.path.isfile(filepath):
        print(f"Fichier introuvable : {filepath}")
        return None, None
    
    df = pd.read_csv(filepath)
    
    # D√©tection du nom de colonne pour les √©missions
    if "emissions" in df.columns:
        emissions_col = "emissions"
    elif "emissions_kg" in df.columns:
        emissions_col = "emissions_kg"
    else:
        raise ValueError(f"Aucune colonne 'emissions' ou 'emissions_kg' dans {filepath}")
    
    # Gestion du temps (optionnel mais propre)
    if "timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
        df = df.sort_values("timestamp")
    
    return df, emissions_col

df_mob, col_mob = load_emissions_csv(file_mobilenet)
df_res, col_res = load_emissions_csv(file_resnet)

print("MobileNetV2 colonnes :", None if df_mob is None else df_mob.columns.tolist())
print("ResNet18 colonnes    :", None if df_res is None else df_res.columns.tolist())
```

- **Estimation de l‚Äôempreinte carbone en phase d‚Äôinf√©rence**
  
Ce bloc estime l‚Äôimpact environnemental r√©el des mod√®les une fois d√©ploy√©s.  
En supposant 3 utilisations par semaine, nous extrapolons la quantit√© de CO‚ÇÇ √©mise sur une ann√©e compl√®te pour MobileNetV2 et ResNet18, afin d‚Äô√©valuer lequel est le plus durable en production.


```python
import matplotlib.pyplot as plt

# Hypoth√®se : 3 utilisations par semaine
uses_per_week = 3
weeks_per_year = 52
n_calls_per_year = uses_per_week * weeks_per_year
print(f"Nombre d'inf√©rences par an : {n_calls_per_year}")

KG_TO_MG = 1_000_000  # 1 kg = 1e6 mg

models = []
yearly_totals_mg = []

# MobileNetV2
if df_mob is not None:
    # Approximation : une "mesure" de CodeCarbon ‚âà co√ªt carbone d'une inf√©rence
    co2_per_call_mob_kg = df_mob[col_mob].mean()
    co2_year_mob_kg = co2_per_call_mob_kg * n_calls_per_year
    co2_year_mob_mg = co2_year_mob_kg * KG_TO_MG

    models.append("MobileNetV2")
    yearly_totals_mg.append(co2_year_mob_mg)

    print(f"MobileNetV2 ~ {co2_year_mob_mg:.2f} mg CO‚ÇÇ / an")

# ResNet18
if df_res is not None:
    co2_per_call_res_kg = df_res[col_res].mean()
    co2_year_res_kg = co2_per_call_res_kg * n_calls_per_year
    co2_year_res_mg = co2_year_res_kg * KG_TO_MG

    models.append("ResNet18")
    yearly_totals_mg.append(co2_year_res_mg)

    print(f"ResNet18 ~ {co2_year_res_mg:.2f} mg CO‚ÇÇ / an")

# --- Barplot annuel en mg ---
plt.figure(figsize=(6, 5))
plt.bar(models, yearly_totals_mg)
plt.ylabel("CO‚ÇÇ annuelle en inf√©rence (mg)")
plt.title("Empreinte carbone annuelle en inf√©rence\n(3 utilisations par semaine)")
plt.tight_layout()
plt.show()
```
